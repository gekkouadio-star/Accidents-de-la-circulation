```python
# ============================================================
# 1. Librairies
# ============================================================
import sys
import time
import pandas as pd
import numpy as np
import matplotlib.cm as cm 
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Patch

# Installation de XGBoost si nécessaire
# !{sys.executable} -m pip install xgboost

# Scikit-Learn : Prétraitement et Pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Scikit-Learn : Modèles de Machine Learning
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier

# Scikit-Learn : Réseaux de Neurones (Alternative stable à TensorFlow)
from sklearn.neural_network import MLPClassifier

# Métriques d'évaluation
from sklearn.metrics import (
    accuracy_score, f1_score, precision_score, recall_score, 
    classification_report, confusion_matrix, ConfusionMatrixDisplay
)

print("✓ Environnement configuré et bibliothèques chargées.")
```

# Chargement des Données

```python
# ============================================================
# CHARGEMENT DES DONNÉES
# ============================================================
# Chemin mis à jour selon votre configuration
df_final = pd.read_csv('./../data_bases/base_complete.csv', sep=';', encoding='utf-8-sig')

print(f"Base : {df_final.shape[0]:,} accidents x {df_final.shape[1]} colonnes")

# Variable cible : GRAV (Gravité de l'accident)
TARGET = 'grav'

print(f"\nVariable cible : {TARGET}")
# Pour la classification, on regarde la répartition des classes
print(df_final[TARGET].value_counts(normalize=True).sort_index() * 100)

# Signification pour rappel : 
# 1: Indemne | 2: Tué | 3: Blessé hospitalisé | 4: Blessé léger
```

# Sélection des Features

```python
# ============================================================
# DÉFINITION DES FEATURES (Accidents de Circulation)
# ============================================================

# Numériques : indicateurs de flux, contraintes et expérience
FEAT_NUM = ['nbv', 'vma', 'an_nais', 'heure']

# Catégorielles : contexte environnemental, technique et infrastructure
# Équivalents : Cause -> atm, Énergie -> motor, Outil -> catv, Site -> lum
FEAT_CAT = ['atm', 'motor', 'catv', 'catr', 'lum', 'surf', 'trajet']

# Binaires : Variables d'état (Équivalent à vos 6 TOP garanties)
# On utilise des variables à deux états pour marquer les conditions spécifiques
FEAT_BIN = ['sexe', 'agg', 'int', 'prof', 'plan', 'situ']

ALL_FEATURES = FEAT_NUM + FEAT_CAT + FEAT_BIN

print(f"Features numériques    : {len(FEAT_NUM)}")
print(f"Features catégorielles : {len(FEAT_CAT)}")
print(f"Features binaires      : {len(FEAT_BIN)}")
print(f"Total                  : {len(ALL_FEATURES)}")
```

# Prétraitement sur mesure (Pipeline)

```python
# ============================================================
# PIPELINES DE PRÉTRAITEMENT
# ============================================================

# 1. Pipeline Numérique : On gère les valeurs aberrantes (ex: vma à 0 ou 999)
num_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')), # Remplace les NaN par la médiane
    ('scaler', StandardScaler())                  # Centre et réduit
])

# 2. Pipeline Catégoriel : Gestion des codes "Inconnus"
# Dans vos données, -1 ou 0 signifie souvent "Inconnu"
cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value=-1)), 
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# 3. Assemblage du ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, FEAT_NUM),
        ('cat', cat_transformer, FEAT_CAT),
        ('bin', cat_transformer, FEAT_BIN) # On traite les binaires comme des catégories
    ]
)
```

# Préparation finale des matrices

```python
# ============================================================
# PRÉPARATION PRÉ-SPLIT
# ============================================================
# On s'assure que 'heure' existe avant le split
df_final['heure'] = df_final['hrmn'].str.split(':').str[0].astype(float)
SEED = 42

# ============================================================
# SPLIT TRAIN / TEST
# ============================================================

# On utilise df_final (le nom défini au début)
X = df_final[ALL_FEATURES].copy()

# CIBLE : décalage pour commencer à 0 (1,2,3,4 -> 0,1,2,3)
y = df_final[TARGET].values - 1 

# Split avec STRATIFICATION
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=SEED, stratify=y
)

print(f"Train : {X_train.shape[0]:,} accidents ({X_train.shape[0]/len(X)*100:.0f}%)")
print(f"Test  : {X_test.shape[0]:,} accidents ({X_test.shape[0]/len(X)*100:.0f}%)")

# ============================================================
# PIPELINE DE PREPROCESSING (VERSION RÉDUITE)
# ============================================================

preprocessor = ColumnTransformer(
    transformers=[
        # Numériques : Imputer + Scaler
        ('num', Pipeline([
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', StandardScaler())
        ]), FEAT_NUM),
        
        # Binaires : Réduction à une colonne par variable (0/1)
        ('bin', Pipeline([
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder(drop='if_binary', sparse_output=False))
        ]), FEAT_BIN),
        
        # Catégorielles : Nettoyage des modalités rares (<2%)
        ('cat', Pipeline([
            ('imputer', SimpleImputer(strategy='constant', fill_value=-1)),
            ('onehot', OneHotEncoder(
                handle_unknown='infrequent_if_exist', 
                min_frequency=0.02, 
                sparse_output=False
            ))
        ]), FEAT_CAT),
    ]
)

# Application du preprocessing
X_train_p = preprocessor.fit_transform(X_train)
X_test_p  = preprocessor.transform(X_test)

print(f"\nDimension finale après réduction : {X_train_p.shape[1]} colonnes")
```

# 1 - MACHINE LEARNING

### Classification de la Gravité

```python
# ============================================================
# FONCTIONS D'ÉVALUATION
# ============================================================

def evaluate_sklearn(model, X_tr, y_tr, X_te, y_te, name):
    """
    Entraîne un modèle de classification, prédit et retourne 
    les métriques de performance par classe.
    """
    t0 = time.time()
    model.fit(X_tr, y_tr)
    fit_time = time.time() - t0

    # Prédictions directes (codes de classe 0, 1, 2, 3)
    y_pred_tr = model.predict(X_tr)
    y_pred_te = model.predict(X_te)

    # Calcul des métriques
    # On utilise 'weighted' pour tenir compte du déséquilibre des classes
    return {
        'Modèle': name,
        'Accuracy Train': accuracy_score(y_tr, y_pred_tr),
        'Accuracy Test': accuracy_score(y_te, y_pred_te),
        'F1-Score (Weighted)': f1_score(y_te, y_pred_te, average='weighted'),
        'Precision Test': precision_score(y_te, y_pred_te, average='weighted', zero_division=0),
        'Recall Test': recall_score(y_te, y_pred_te, average='weighted', zero_division=0),
        'Temps (s)': fit_time,
    }

print("✓ Fonctions d'évaluation adaptées à la classification définies")
```

### Entraînement des Modèles ML

```python
# ============================================================
# DÉFINITION DES MODÈLES ML (VERSION CORRIGÉE)
# ============================================================

models_ml = {
    'Régression Logistique': LogisticRegression(
        max_iter=1000, 
        class_weight='balanced', # Crucial pour gérer le peu de "Tués"
        random_state=SEED
    ),
    
    'Random Forest': RandomForestClassifier(
        n_estimators=200, 
        max_depth=12,             # Un peu réduit pour éviter le surapprentissage
        min_samples_leaf=10,
        class_weight='balanced', 
        random_state=SEED, 
        n_jobs=-1
    ),
    
    'Gradient Boosting': GradientBoostingClassifier(
        n_estimators=150,         # Réduit pour gagner du temps au début
        max_depth=5, 
        learning_rate=0.1,
        subsample=0.8, 
        random_state=SEED
    ),
    
    'XGBoost': XGBClassifier(
        n_estimators=200, 
        max_depth=6, 
        learning_rate=0.1,
        subsample=0.8, 
        colsample_bytree=0.8,
        # 'multi:softprob' est souvent automatique, mais on le garde pour la clarté
        objective='multi:softprob', 
        random_state=SEED, 
        n_jobs=-1, 
        verbosity=0
    )
}

print(f"✓ {len(models_ml)} modèles de classification définis (multi_class supprimé)")
```

```python
# ============================================================
# ENTRAÎNEMENT DES MODÈLES ML (CLASSIFICATION)
# ============================================================
results_ml = []
trained_models = {}

for name, model in models_ml.items():
    print(f"► {name}...", end=" ")
    
    # Entraînement et évaluation
    res = evaluate_sklearn(model, X_train_p, y_train, X_test_p, y_test, name)
    
    results_ml.append(res)
    trained_models[name] = model
    
    # Affichage console rapide
    print(f"Acc={res['Accuracy Test']:.2%} | F1={res['F1-Score (Weighted)']:.2%} | t={res['Temps (s)']:.1f}s")

# Création du DataFrame de comparaison
df_ml = pd.DataFrame(results_ml).set_index('Modèle')

print("\n✓ Entraînement ML terminé")

# --- AFFICHAGE STYLISÉ POUR LE RAPPORT ---
def highlight_max(s):
    '''Met en gras et en évidence la valeur maximale de chaque colonne'''
    is_max = s == s.max()
    return ['font-weight: bold; background-color: #e6f3ff' if v else '' for v in is_max]

# On trie par F1-Score et on applique le style
df_styled = df_ml.sort_values(by='F1-Score (Weighted)', ascending=False).style \
    .apply(highlight_max, subset=['Accuracy Test', 'F1-Score (Weighted)']) \
    .format({
        'Accuracy Train': '{:.2%}',
        'Accuracy Test': '{:.2%}',
        'F1-Score (Weighted)': '{:.2%}',
        'Precision Test': '{:.2%}',
        'Recall Test': '{:.2%}',
        'Temps (s)': '{:.1f}s'
    })

display(df_styled)
```

**Commentaire :**  L'entraînement des quatre architectures de classification permet de tirer les conclusions suivantes :

1. Le Champion : XGBoost
Performance : Avec une Accuracy de 60,18% et un F1-Score de 58,34%, XGBoost s'impose comme le modèle le plus robuste pour prédire les catégories d'accidents.

Efficacité : Il offre le meilleur rapport performance/temps en traitant l'ensemble des données en seulement 5,3 secondes.

2. L'anomalie du Gradient Boosting
Nous observons un temps d'exécution particulièrement élevé pour le Gradient Boosting classique (124,5s), soit environ 23 fois plus lent que XGBoost.

Explication technique : Contrairement à XGBoost qui utilise la parallélisation et des histogrammes pour accélérer le calcul, l'implémentation standard de Scikit-Learn est séquentielle. Elle teste chaque seuil de division de manière exhaustive, ce qui la rend inefficace sur des volumes de données importants par rapport aux optimisations modernes.

3. Généralisation et Surapprentissage
L'écart entre l'Accuracy Train (63,2%) et l'Accuracy Test (60,2%) est très réduit (environ 3%).

Cela démontre une excellente capacité de généralisation du modèle : il n'y a pas de surapprentissage (overfitting) marqué, ce qui garantit que le modèle restera fiable sur de nouvelles données d'accidents.

### Résultats Machine Learning

```python
# ============================================================
# TABLEAU DE COMPARAISON ML (ADAPTÉ CLASSIFICATION)
# ============================================================

# Colonnes adaptées à la classification
display_cols_ml = ['Accuracy Train', 'Accuracy Test', 'F1-Score (Weighted)', 'Recall Test', 'Temps (s)']

# On trie par F1-Score (la métrique la plus robuste pour la gravité)
df_ml_sorted = df_ml[display_cols_ml].sort_values('F1-Score (Weighted)', ascending=False)

print("=== COMPARAISON DES MODÈLES ML (CLASSIFICATION) ===\n")
# On affiche les scores en pourcentage pour plus de lisibilité
print((df_ml_sorted * 100).round(1).to_string())

best_ml_name = df_ml_sorted.index[0]
print(f"\n★ Meilleur modèle ML : {best_ml_name}")

# --- Visualisation ---
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Graphique 1 : Accuracy Train vs Test (Capacité de généralisation)
df_plot_acc = df_ml_sorted[['Accuracy Train', 'Accuracy Test']]
df_plot_acc.plot(kind='barh', ax=axes[0], color=['#a8dadc', '#457b9d'], edgecolor='white')
axes[0].set_title('Précision (Accuracy) Train vs Test')
axes[0].set_xlabel('Score')
axes[0].set_xlim(0, 1) # Les scores vont de 0 à 1

# Graphique 2 : F1-Score (Équilibre Précision/Rappel)
df_plot_f1 = df_ml_sorted[['F1-Score (Weighted)']]
df_plot_f1.plot(kind='barh', ax=axes[1], color=['#e76f51'], edgecolor='white')
axes[1].set_title('F1-Score (Moyenne pondérée)')
axes[1].set_xlabel('Score')
axes[1].set_xlim(0, 1)

plt.suptitle('Performance des Modèles de Classification de Gravité', fontsize=15, y=1.05)
plt.tight_layout()
plt.show()
```

**Commentaire :** 

L'évaluation comparative des quatre algorithmes de Machine Learning pour la prédiction de la gravité des accidents permet de tirer les conclusions suivantes :

1. Suprématie de l'approche par Boosting
Les modèles basés sur le gradient boosting dominent nettement le classement :

XGBoost arrive en tête avec la meilleure Accuracy Test (60,2%) et le meilleur F1-Score (58,3%).

Gradient Boosting affiche des scores quasiment identiques, confirmant que les algorithmes de "boosting" sont les plus adaptés à la structure complexe de nos données routières.

2. Efficacité Computationnelle : Le point critique
L'écart de temps d'exécution est le facteur le plus frappant de cette comparaison :

XGBoost est extrêmement performant avec un temps de traitement de 514 ms (environ 0,5 seconde).

Gradient Boosting prend 12 071 ms (soit environ 12 secondes).

Analyse : Bien que les scores de précision soient proches, XGBoost est environ 23 fois plus rapide. Cela s'explique par sa gestion optimisée de la mémoire et la parallélisation des calculs, contrairement au Gradient Boosting standard qui est purement séquentiel.

3. Diagnostic de la Généralisation
En comparant l'Accuracy Train (63,2%) et l'Accuracy Test (60,2%) de XGBoost :

L'écart est très faible (3%), ce qui indique une excellente généralisation.

Le modèle n'est pas en situation de surapprentissage (overfitting) significatif. Il a appris des motifs réels et non du "bruit" statistique, ce qui garantit sa fiabilité sur de nouvelles données.

4. Analyse du F1-Score vs Accuracy
On remarque que le F1-Score (58,3%) est légèrement inférieur à l'Accuracy.

Cela reflète la difficulté classique du déséquilibre des classes : le modèle est très précis sur les classes majoritaires (ex: accidents sans gravité), mais plus hésitant sur les cas rares mais critiques (accidents graves).

*Conclusion :* Le modèle XGBoost est sélectionné comme le "champion" final. Il offre non seulement la meilleure capacité de détection, mais aussi une vitesse de calcul indispensable pour une mise en production ou des analyses répétées sur de grands volumes de données.

# 2 - DEEPL LEARNING

### Réseaux de Neurones

```python
# On définit les modèles de réseaux de neurones
dl_models_sklearn = {
    'MLP Simple': MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500, random_state=42),
    'MLP Deep': MLPClassifier(hidden_layer_sizes=(128, 64, 32), max_iter=500, random_state=42)
}

results_dl = {}

print("=== ENTRAÎNEMENT DES RÉSEAUX DE NEURONES (MODE SÉCURISÉ) ===")
for name, model in dl_models_sklearn.items():
    print(f"► {name}...", end=" ", flush=True)
    
    # Entraînement
    model.fit(X_train_p, y_train)
    
    # Prédiction
    y_pred = model.predict(X_test_p)
    
    results_dl[name] = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'F1-Score': f1_score(y_test, y_pred, average='weighted')
    }
    print(f"Terminé (Acc: {results_dl[name]['Accuracy']:.2%})")

# Affichage final

print("\n", pd.DataFrame(results_dl).T)
```

**Commentaire :** 

Les résultats montrent une dynamique intéressante par rapport aux modèles de Machine Learning classiques :

Le réseau "Simple" l'emporte : *MLP Simple* (58,2% Accuracy) surpasse *le MLP Deep* (54,6%).

Dans le cas de nos données tabulaires sur les accidents, augmenter la profondeur du réseau (ajouter des couches) n'a pas amélioré la précision. Au contraire, cela semble avoir introduit une complexité inutile qui dégrade la performance sur le jeu de test.

### Entraînement et Évaluation du Modèles DL

```python
# ============================================================
# ARCHITECTURES RÉSEAUX DE NEURONES
# ============================================================

# On définit les modèles selon ta logique de profondeur
dl_models_sklearn = {
    'MLP Simple': MLPClassifier(
        hidden_layer_sizes=(64, 32), 
        max_iter=500, 
        early_stopping=True, 
        random_state=42
    ),
    'MLP Deep': MLPClassifier(
        hidden_layer_sizes=(128, 64, 32), 
        max_iter=500, 
        early_stopping=True, 
        random_state=42
    )
}

results_dl = {}

print("=== ENTRAÎNEMENT DES RÉSEAUX DE NEURONES (MODE SÉCURISÉ) ===")

for name, model in dl_models_sklearn.items():
    print(f"► {name} en cours...", end=" ", flush=True)
    
    t0 = time.time()
    
    # 1. Entraînement
    model.fit(X_train_p, y_train)
    fit_time = time.time() - t0
    
    # 2. Prédiction
    y_pred = model.predict(X_test_p)
    y_pred_train = model.predict(X_train_p)
    
    # 3. Stockage des résultats (Même structure que ton code initial)
    results_dl[name] = {
        'Modèle': name,
        'Accuracy Train': accuracy_score(y_train, y_pred_train),
        'Accuracy Test': accuracy_score(y_test, y_pred),
        'F1-Score (Weighted)': f1_score(y_test, y_pred, average='weighted'),
        'Temps (s)': fit_time,
        'Iterations': model.n_iter_,
        'model_object': model # Pour garder une trace si besoin
    }
    
    print(f"OK | Acc Test: {results_dl[name]['Accuracy Test']:.2%} | t={fit_time:.1f}s")

# ============================================================
# TABLEAU FINAL POUR L'ÉVALUATION
# ============================================================

# Transformation en DataFrame pour un affichage propre dans ton rapport
df_dl_final = pd.DataFrame([
    {k: v for k, v in res.items() if k != 'model_object'} 
    for res in results_dl.values()
]).set_index('Modèle')

print("\n--- COMPARAISON DES ARCHITECTURES NEURONALES ---")
display(df_dl_final.sort_values('F1-Score (Weighted)', ascending=False))
```

**Commentaire :**

MLP Simple (59,48% Accuracy) vs MLP Deep (59,38% Accuracy).

Les deux architectures produisent des résultats quasi-identiques. Ajouter de la profondeur (MLP Deep) n'a pas permis de capturer plus de complexité, mais a augmenté le temps de calcul de 50% (de 8,4s à 12,2s).

### Courbes d'Apprentissage du Modèle DL

```python
# ============================================================
# COURBES DE LOSS : ENTRAÎNEMENT (Version Scikit-Learn)
# ============================================================

# On adapte le nombre de sous-graphiques au nombre de modèles (ici 2 : Simple et Deep)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, (name, res) in zip(axes.flat, results_dl.items()):
    # Récupération du modèle entraîné
    model = res['model_object']
    
    # Scikit-learn stocke la perte dans loss_curve_
    loss_values = model.loss_curve_
    epochs_range = range(1, len(loss_values) + 1)

    ax.plot(epochs_range, loss_values, label='Train Loss', color='steelblue', linewidth=2)
    
    # Note : MLPClassifier avec early_stopping=True stocke aussi 
    # les scores de validation dans validation_scores_ (si activé)
    if hasattr(model, 'validation_scores_') and model.validation_scores_:
        # On convertit le score de précision en "perte" approximative pour le visuel
        val_loss_simulated = [1 - s for s in model.validation_scores_]
        ax.plot(range(1, len(val_loss_simulated) + 1), val_loss_simulated, 
                label='Validation Error', color='salmon', linestyle='--')

    ax.set_title(f"{name}\n(Convergence à l'itération {res['Iterations']})")
    ax.set_xlabel('Itérations')
    ax.set_ylabel('Loss (Log-Loss)')
    ax.legend()
    ax.grid(True, alpha=0.3)

    # Annoter la fin de l'entraînement
    final_loss = loss_values[-1]
    ax.annotate(f'Final Loss:\n{final_loss:.4f}',
                xy=(len(loss_values), final_loss), 
                xytext=(len(loss_values)*0.7, final_loss*1.2),
                arrowprops=dict(facecolor='black', shrink=0.05, width=1),
                bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.suptitle('Courbes d\'Apprentissage — Réseaux de Neurones (MLP)', fontsize=15, y=1.05)
plt.tight_layout()
plt.show()
```

# 3 - COMPARAISON ML/DL (prédiction)

```python
# ============================================================
# TABLEAU COMPARATIF GLOBAL : ML vs DL (VERSION CAUSES)
# ============================================================

# 1. Extraction des résultats DL (Scikit-Learn)
dl_rows = []
for name, res in results_dl.items():
    dl_rows.append({
        'Modèle': name,
        'Acc Train': res['Accuracy Train'],
        'Acc Test': res['Accuracy Test'],
        'F1-Score Test': res['F1-Score (Weighted)'],
        'Temps (s)': res['Temps (s)'],
        'Type': 'DL'
    })

# 2. Extraction des résultats ML (On suppose que ton df_ml contient déjà Acc/F1)

ml_rows = []
for name in df_ml.index:
    ml_rows.append({
        'Modèle': name,
        'Acc Train': df_ml.loc[name, 'Accuracy Train'],
        'Acc Test': df_ml.loc[name, 'Accuracy Test'],
        'F1-Score Test': df_ml.loc[name, 'F1-Score (Weighted)'],
        'Temps (s)': df_ml.loc[name, 'Temps (s)'],
        'Type': 'ML'
    })

# 3. Fusion des deux mondes
df_all = pd.concat([
    pd.DataFrame(ml_rows).set_index('Modèle'),
    pd.DataFrame(dl_rows).set_index('Modèle')
]).sort_values('F1-Score Test', ascending=False) # On trie par le meilleur score

# Affichage
print("=" * 85)
print("COMPARAISON GLOBALE — Machine Learning vs Deep Learning")
print("=" * 85)
print(df_all.round(3).to_string())

best_name = df_all.index[0]
best_type = df_all.loc[best_name, 'Type']

print(f"\n★ MEILLEUR MODÈLE GLOBAL : {best_name}")
print(f"  Type      : {best_type}")
print(f"  F1-Score  : {df_all.loc[best_name, 'F1-Score Test']:.2%}")
print(f"  Accuracy  : {df_all.loc[best_name, 'Acc Test']:.2%}")
print("=" * 85)
```

**Commentaire :**

Le modèle XGBoost est meilleur comme le moteur de prédiction le montre. Il offre la meilleure précision, la plus grande rapidité d'exécution et une interprétabilité claire grâce à l'analyse des variables déterminantes. Ce projet démontre que la maîtrise du prétraitement des données (One-Hot Encoding, Scaling) associée à un algorithme de Boosting performant permet de modéliser la gravité des accidents routiers.

```python
# ============================================================
# VISUALISATION COMPARATIVE FINALE : ML vs DL
# ============================================================

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 1. Comparaison de l'ACCURACY (Précision globale)
df_plot_acc = df_all.sort_values('Acc Test', ascending=True)
axes[0].barh(df_plot_acc.index, df_plot_acc['Acc Test'], 
             color=['steelblue' if t == 'ML' else 'coral' for t in df_plot_acc['Type']],
             edgecolor='white')
axes[0].set_title('Précision Globale (Accuracy Test)', fontsize=13)
axes[0].set_xlabel('Score (%)')
axes[0].set_xlim(0, 1.0) # Échelle de 0 à 100%

# 2. Comparaison du F1-SCORE (Équilibre Précision/Rappel)
df_plot_f1 = df_all.sort_values('F1-Score Test', ascending=True)
axes[1].barh(df_plot_f1.index, df_plot_f1['F1-Score Test'], 
             color=['steelblue' if t == 'ML' else 'coral' for t in df_plot_f1['Type']],
             edgecolor='white')
axes[1].set_title('Qualité de la Classification (F1-Score)', fontsize=13)
axes[1].set_xlabel('F1-Score (Weighted)')
axes[1].set_xlim(0, 1.0)

# Légende personnalisée
legend_elements = [Patch(facecolor='steelblue', label='Machine Learning (ML)'),
                   Patch(facecolor='coral', label='Deep Learning (DL)')]
fig.legend(handles=legend_elements, loc='upper center', bbox_to_anchor=(0.5, 0.95), 
           ncol=2, fontsize=12)

plt.suptitle('Performance des modèles pour la prédiction des causes d\'accidents', 
             fontsize=16, y=1.05)
plt.tight_layout()
plt.show()
```

### Choix du Modèle final

```python
# ============================================================
# ANALYSE DU MEILLEUR MODÈLE (CORRIGÉ & LISIBLE)
# ============================================================
best_global_name = df_all.index[0]
best_global_type = df_all.loc[best_global_name, 'Type']

print(f"Analyse du Champion : {best_global_name} ({best_global_type})")
print(f"{'=' * 50}")

# --- 1. Récupération des prédictions ---
if best_global_type == 'ML':
    best_model = trained_models[best_global_name]
    y_pred_final = best_model.predict(X_test_p)
else:
    best_dl_obj = results_dl[best_global_name]['model_object']
    y_pred_final = best_dl_obj.predict(X_test_p)

# --- 2. Définition des noms des classes ---
# REMPLACE par tes vrais noms de catégories (ex: 'Indemne', 'Tué', etc.)
class_names = ['Indemne', 'Blessé Léger', 'Blessé Grave', 'Tué'] 

# --- 3. Matrice de Confusion ---
fig, axes = plt.subplots(1, 2, figsize=(18, 7))

# Rapport textuel (plus précis pour le rapport)
print(classification_report(y_test, y_pred_final, target_names=class_names))

cm = confusion_matrix(y_test, y_pred_final)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
disp.plot(ax=axes[0], cmap='Blues', values_format='d', xticks_rotation=45)
axes[0].set_title(f'Matrice de Confusion : {best_global_name}')

# --- 4. Importance des Variables avec NOMS REELS ---
if best_global_type == 'ML' and hasattr(best_model, 'feature_importances_'):
    # On récupère les noms des colonnes après le preprocessing (75 colonnes)
    # Note: Adapte 'preprocessor' au nom de ton objet de transformation
    feature_names = preprocessor.get_feature_names_out() 
    
    importances = best_model.feature_importances_
    indices = np.argsort(importances)[-10:] # Top 10
    
    axes[1].barh(range(len(indices)), importances[indices], color='steelblue')
    axes[1].set_yticks(range(len(indices)))
    # C'est ici qu'on remplace les chiffres par les noms des colonnes
    axes[1].set_yticklabels([feature_names[i] for i in indices])
    axes[1].set_title('Top 10 des facteurs déterminants l\'accident')
    axes[1].set_xlabel('Score d\'importance')
else:
    # Si DL, on affiche la distribution nommée
    counts = pd.Series(y_pred_final).map({i: name for i, name in enumerate(class_names)}).value_counts()
    counts.plot(kind='bar', ax=axes[1], color='coral')
    axes[1].set_title('Distribution des causes prédites par l\'IA')
    axes[1].set_ylabel('Nombre de dossiers')

plt.tight_layout()
plt.show()
```

```python
# ============================================================
# PRÉDICTION PAR CATÉGORIE D'ACCIDENTS
# ============================================================
# Calcul de la précision (Recall) pour chaque classe
# Le "Recall" mesure la capacité du modèle à trouver tous les cas d'une classe
class_names = ['Indemne', 'Blessé Léger', 'Blessé Grave', 'Tué']
recalls = recall_score(y_test, y_pred_final, average=None)

perf_by_class = pd.DataFrame({
    'Catégorie': class_names,
    'Taux de Détection (%)': (recalls * 100).round(1)
}).set_index('Catégorie')

print("=== Prédiction par cétogorie d'accidents ===\n")
print(perf_by_class.to_string())

# --- Visualisation ---
fig, ax = plt.subplots(figsize=(10, 5))

# Création du graphique à barres
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'] # Couleurs distinctes
perf_by_class['Taux de Détection (%)'].plot(kind='bar', ax=ax, color=colors, edgecolor='white')

# Esthétique
ax.set_title(f'Capacité de détection du modèle : {best_global_name}', fontsize=14)
ax.set_ylabel('Précision (Recall) en %')
ax.set_ylim(0, 100)
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=0)

# Ajout des valeurs au-dessus des barres
for i, v in enumerate(perf_by_class['Taux de Détection (%)']):
    ax.text(i, v + 2, f"{v}%", ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

**Commentaire :**

Ce graphique illustre la capacité du modèle XGBoost à identifier correctement chaque catégorie de gravité. Les résultats sont contrastés et riches en enseignements :

1. Performance sur les extrêmes (Indemne et Tué)

**Indemne (77,9%) :** Le modèle identifie très bien les accidents sans gravité. C'est la classe la plus représentée, ce qui facilite l'apprentissage.

**Tué (57,1%) :** Le score est encourageant. Le modèle parvient à détecter plus d'un accident mortel sur deux, ce qui suggère que les accidents mortels possèdent des caractéristiques (signatures) très spécifiques (vitesse élevée, choc frontal, etc.).

2. Le défi des classes intermédiaires

**Blessé Grave (29,7%) :** La détection chute. Ces accidents partagent probablement beaucoup de caractéristiques communes avec les accidents mortels ou légers, rendant la frontière floue pour le modèle.

**Blessé Léger (2,5%) :** C'est le point critique. Le modèle échoue presque totalement à identifier cette classe.

**Commentaire :**

Top 15 des Facteurs Déterminants" nous permet de comprendre ce qui influence réellement la gravité d'un accident selon le modèle XGBoost. Contrairement aux modèles statistiques classiques, l'importance par permutation mesure l'impact réel de chaque variable sur la précision finale.

1. Le facteur dominant : **catv (Catégorie de véhicule)**
Impact : 0.131

C'est, de loin, la variable la plus importante. Cela signifie que le type de véhicule (poids lourd, deux-roues, voiture légère) est le prédicteur n°1 de la gravité.

Analyse : Un choc impliquant un usager vulnérable (moto/vélo) ou un véhicule de fort tonnage change radicalement l'issue de l'accident.

2. Profil de l'usager et contexte : an_nais et trajet
**an_nais (0.027) :** L'âge du conducteur/usager arrive en deuxième position. La fragilité physique liée à l'âge ou le manque d'expérience des plus jeunes sont des facteurs clés.

**trajet (0.022) :** Le motif du déplacement (travail, loisirs) influence les comportements au volant et donc la gravité potentielle.

3. Facteurs environnementaux et infrastructure
On remarque que des variables comme **vma (vitesse maximale autorisée)**, **catr (catégorie de route)** et **heure** ont un impact modéré mais réel (entre 0.007 et 0.009).

Curieusement, **les facteurs météo (atm)** et **l'état de la surface (surf)** apparaissent en bas de classement (0.001).

Interprétation : Cela suggère que les conducteurs adaptent souvent leur prudence à la pluie, mais qu'ils sont surpris par des facteurs structurels (type de route) ou liés au véhicule.
